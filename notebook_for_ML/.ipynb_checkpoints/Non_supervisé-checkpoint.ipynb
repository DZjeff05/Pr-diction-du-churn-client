#### appliquons notre model

```python
# Min-Max
minmax_cols = ['Satisfaction_Score']

# Standardisation
standard_cols = [
    'Annual_Income',
    'Total_Spend',
    'Years_as_Customer',
    'Num_of_Purchases',
    'Average_Transaction_Amount',
    'Num_of_Returns',
    'Num_of_Support_Contacts',
    'Last_Purchase_Days_Ago',
    'Age'
]

# Catégorielles
categorical_cols = ['Gender', 'Promotion_Response']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("minmax", MinMaxScaler(), minmax_cols),
        ("standard", StandardScaler(), standard_cols),
        ("cat", OneHotEncoder(drop=None, handle_unknown="ignore"), categorical_cols)
    ]
)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X_preprocessed = preprocessor.fit_transform(
    df[minmax_cols + standard_cols + categorical_cols]
)

inertia = []

for k in range(2, 10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_preprocessed)
    inertia.append(km.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(range(2, 10), inertia, marker='o')
plt.xlabel("Nombre de clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
cluster_pipeline = Pipeline(steps=[
    ("preprocessing", preprocessor),
    ("kmeans", KMeans(n_clusters=3, random_state=42))
])
```

```python
from sklearn import set_config
set_config(display="diagram")

cluster_pipeline
```

```python
X = df[minmax_cols + standard_cols + categorical_cols]

cluster_pipeline.fit(X)

clusters = cluster_pipeline.predict(X)
pd.Series(clusters).value_counts()
```

```python
CLUSTER_NAMES = {
    0: "Clients Fidèles",
    1: "Clients à Risque",
    2: "Clients Volatils"
}

df["Cluster"] = clusters
df["Cluster_Label"] = df["Cluster"].map(CLUSTER_NAMES)

df[["Cluster", "Cluster_Label"]].head()
```

```python
df.groupby("Cluster_Label")[[
    'Total_Spend',
    'Num_of_Purchases',
    'Years_as_Customer',
    'Satisfaction_Score',
    'Last_Purchase_Days_Ago'
]].mean()
```
